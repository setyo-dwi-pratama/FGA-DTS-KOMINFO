# **Persiapan Dataset**

```
import pandas as pd
import numpy as np
```

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
```

```
df = pd.read_csv(url, header=None)
df.head()
```

```
headers = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style',
           'drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type',
           'num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horspower','peak-rpm',
           'city-mpg','highway-mpg','price']
```

```
df.columns=headers
df.head()
```

```
df.info()
```

```
# Menggantikan nilai pada dataset dari '?' dengan NaN, sehingga terbaca sebagai missing value
df.replace("?", np.nan, inplace=True)
df.head()
```

```
# Melihat missing value berdasarkan True dan False
missing_data = df.isnull()
missing_data.head(5)
```

```
# Melihat kolom-kolom yang terdapat missing value
for column in missing_data.columns.values.tolist():
  print("Nama Kolom", column)
  print(missing_data[column].value_counts())
  print("")
```

Sehingga dari 205 baris terdapat 7 kolom yang memiliki missing value, diantaranya:
1. "normalized-losses": 41 missing data
2. "num-of-doors": 2 missing data
3. "bore": 4 missing data
4. "stroke" : 4 missing data
5. "horsepower": 2 missing data
6. "peak-rpm": 2 missing data
7. "price": 4 missing data

# **Cleaning Dataset**

## **Kolom normalized-losses** 
Mengganti dengan nilai rata-rata

```
#Konversi type data float
df['normalized-losses'] = df['normalized-losses'].astype(float)
```

```
rata = df['normalized-losses'].mean()
```

```
rata
```

```
df['normalized-losses'] = df['normalized-losses'].replace(np.nan, rata)
```

```
df
```
