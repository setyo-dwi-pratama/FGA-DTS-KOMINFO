```
import pandas as pd
import numpy as np
```

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
```

```
df = pd.read_csv(url, header=None)
df.head()
```

```
headers = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style',
           'drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type',
           'num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horspower','peak-rpm',
           'city-mpg','highway-mpg','price']
```

```
df.columns=headers
df.head()
```

```
df.info()
```

```
df['normalized-losses'].value_counts()
```

```
df['normalized-losses'] = df['normalized-losses'].str.replace('?','0')
```

```
df['normalized-losses'] = df['normalized-losses'].astype(int)
```

```
rata = df['normalized-losses'].mean()
```

```
rata
```

```
df['normalized-losses'] = df['normalized-losses'].replace(0, rata)
```
